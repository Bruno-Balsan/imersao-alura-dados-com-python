```python
df.columns = [
    'ano',
    'senerioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'usd',
    'residencia',
    'remoto',
    'empresa',
    'tamanho_empresa'
]

print("Colunas traduzidas com sucesso! Novas colunas:")
print(df.columns)
```

```python
df['senerioridade'] = df['senerioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Médio',
    'EN': 'Júnior',
    'EX': 'Executivo'
})

print("Categorias de 'senerioridade' traduzidas com sucesso! Novas contagens:")
print(df['senerioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

print("Categorias de 'contrato' traduzidas com sucesso! Novas contagens:")
print(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})

print("Categorias de 'tamanho_empresa' traduzidas com sucesso! Novas contagens:")
print(df['tamanho_empresa'].value_counts())
```

```python
print("Primeiras linhas do DataFrame com as traduções confirmadas:")
df.head(10)
```

```python
df['cargo'] = df['cargo'].replace({
    'Data Scientist': 'Cientista de Dados',
    'Software Engineer': 'Engenheiro de Software',
    'Data Engineer': 'Engenheiro de Dados',
    'Data Analyst': 'Analista de Dados',
    'Engineer': 'Engenheiro'
})

print("Categorias de 'cargo' traduzidas com sucesso! Novas contagens (top 10):")
print(df['cargo'].value_counts().head(10))
```
