```python
df['senerioridade'] = df['senerioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Médio',
    'EN': 'Júnior',
    'EX': 'Executivo'
})

print("Categorias de 'senerioridade' traduzidas com sucesso! Novas contagens:")
print(df['senerioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

print("Categorias de 'contrato' traduzidas com sucesso! Novas contagens:")
print(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
})

print("Categorias de 'tamanho_empresa' traduzidas com sucesso! Novas contagens:")
print(df['tamanho_empresa'].value_counts())
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    "nome":['Bruno','Emily',"Ana",'Erick', 'Valkiria', 'Robson'],
    "salario":[4000, np.nan, 5000, np.nan, 100000, 5700]
})

#calcula a média e substitui os nulos e arredonda
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#Calcula a mediana e substitui os nulos
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "dia":['segunda','terça','quarta','quinta','sexta','sabado','domingo'],
    "temperatura":[30, np.nan, np.nan, 28, 27, 25, 26]
})
#pega a temperatura do dia anterior
df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "dia":['segunda','terça','quarta','quinta','sexta','sabado','domingo'],
    "temperatura":[30, np.nan, np.nan, 28, 27, 25, 26]
})
#pega a temperatura do dia seguinte
df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nome":['Bruno','Emily',"Ana",'Erick', 'Valkiria', 'Robson'],
    "cidade":['Osasco', 'São Paulo', np.nan, 'Curitiba', np.nan, np.nan]
})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("cidade não informada")
df_cidades
```
